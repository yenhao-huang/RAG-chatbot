```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Embedding Model": ["sentence-transformer", "Alibaba"],
    "Precision@5 (Retrieval)": [37, 80],
    "LLM-as-a-Judge (QA)": [54, 76]
}

df = pd.DataFrame(data)

# 只取 Precision@5
metric = "Precision@5 (Retrieval)"
models = df["Embedding Model"].tolist()
scores = df[metric].tolist()

x = np.arange(len(models))
bar_width = 0.5

plt.figure(figsize=(5, 5))

# 同色系（淺藍 & 深藍）
colors = ["#6baed6", "#2171b5"]

plt.bar(x, scores, width=bar_width, color=colors, tick_label=models)

plt.ylabel("Percentage (%)")
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt

# 資料
methods = [
    "No Optimization",
    "Quant. + LoRA",
    "Quant. + LoRA + SeqLen"
]
memory_usage = [23.5, 2.7, 1.5]

# 藍色同色系
colors = ["#2171b5", "#6baed6", "#9ecae1"]

# 繪圖
plt.figure(figsize=(6, 4))
bars = plt.bar(methods, memory_usage, color=colors)

# 標註數值
for bar, value in zip(bars, memory_usage):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
             f"{value} GB", ha="center", va="bottom", fontsize=10)

plt.ylabel("GPU Memory Usage (GB)")
plt.ylim(0, 25)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
