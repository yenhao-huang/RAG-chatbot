```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 整合所有模型數據
data = {
    "Embedding Model": [
        "Sentence-Transformer",
        "Qwen3-0.6B",
        "Alibaba-NLP",
        "BGE-M3",
        "BGE-Large",
        "Conan-Embedding-V1"
    ],
    "Precision@5 (Retrieval)": [27, 81.8, 81.8, 85.5, 80, 80],
}

df = pd.DataFrame(data)

metric = "Precision@5 (Retrieval)"
models = df["Embedding Model"].tolist()
scores = df[metric].tolist()

x = np.arange(len(models))
bar_width = 0.6

plt.figure(figsize=(8, 6))

# 用淺藍到深藍的漸層
colors = ["#c6dbef", "#9ecae1", "#6baed6", "#4292c6", "#2171b5", "#084594"]

bars = []
for i, (label, value, color) in enumerate(zip(models, scores, colors)):
    bar = plt.bar(i, value, width=bar_width, color=color)
    bars.append(bar)

# 移除 x 軸標籤
plt.xticks([])

# 在柱子上方加數字
for i, value in enumerate(scores):
    plt.text(i, value + 1, f"{value:.1f}%", ha="center", va="bottom", fontsize=9)

plt.ylabel("Precision@5 (%)")
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.6)

# 把 legend 放在圖的上方，橫向排列
plt.legend([b[0] for b in bars], models, loc="upper center",
           bbox_to_anchor=(0.5, 1.12), ncol=3, frameon=False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 資料
methods = [
    "No Optimization",
    "Quant. + LoRA",
    "Quant. + LoRA + SeqLen"
]
memory_usage = [23.5, 2.7, 1.5]

# 藍色同色系
colors = ["#2171b5", "#6baed6", "#9ecae1"]

# 繪圖
plt.figure(figsize=(6, 4))
bars = plt.bar(methods, memory_usage, color=colors)

# 標註數值
for bar, value in zip(bars, memory_usage):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
             f"{value} GB", ha="center", va="bottom", fontsize=10)

plt.ylabel("GPU Memory Usage (GB)")
plt.ylim(0, 25)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
